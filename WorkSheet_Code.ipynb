14.	Write a python program to check whether a webpage contains a title or not.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
req_url = 'https://www.axisbank.com/'
```

```python
res_ = requests.get(url=req_url)
```

```python
res_
```

```python
scrap_Content = BeautifulSoup(res_.content)
scrap_Content
```

```python
title = scrap_Content.find_all("title")
```

```python
len(title)
```

```python
if len(title) > 0:
    print('Title is Present')
```
